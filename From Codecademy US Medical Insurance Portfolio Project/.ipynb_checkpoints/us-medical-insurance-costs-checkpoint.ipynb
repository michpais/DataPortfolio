# U.S. Medical Insurance Costs

Scope of the project: to see what region has the healthiest BMI, least smokers, and lowest insurance cost. Then, it is to figure out what factor has the biggest effect on BMI

Note: A normal BMI, according to the CDC, is 18.5 to 24.9. The middle of the range, a BMI of 21.75, will be used to determine "healthiest" bmi

```python
# import the modules that will be needed for the analysis
import csv
import matplotlib.pyplot as plt
```

```python
def average(user_list):
    avg = sum(user_list)/len(user_list)
    return avg
```

```python
# create list of lists that hold insurance information
insurance_data = []

# fill in the list of lists that hold insurance information
with open('insurance.csv') as insurance_csv:
    insurance_reader = csv.reader(insurance_csv, delimiter=',')
    for person in insurance_reader:
        #print(person)
        insurance_data.append(person)
# remove the first list, which were the header names in the csv file
insurance_data = insurance_data[1:]
```

```python
sw_bmi_total = 0; sw_count = 0; sw_smoker_count = 0; sw_ins_cost = 0
se_bmi_total = 0; se_count = 0; se_smoker_count = 0; se_ins_cost = 0
ne_bmi_total = 0; ne_count = 0; ne_smoker_count = 0; ne_ins_cost = 0
nw_bmi_total = 0; nw_count = 0; nw_smoker_count = 0; nw_ins_cost = 0


for person in insurance_data:
    if person[5] == 'southwest':
        sw_bmi_total += float(person[2])
        sw_count += 1
        sw_ins_cost += float(person[6])
        if person[4] == 'yes':
            sw_smoker_count += 1
    elif person[5] == 'southeast':
        se_bmi_total += float(person[2])
        se_count += 1
        se_ins_cost += float(person[6])
        if person[4] == 'yes':
            se_smoker_count += 1
    elif person[5] == 'northwest':
        nw_bmi_total += float(person[2])
        nw_count += 1
        nw_ins_cost += float(person[6])
        if person[4] == 'yes':
            nw_smoker_count += 1
    elif person[5] == 'northeast':
        ne_bmi_total += float(person[2])
        ne_count += 1
        ne_ins_cost += float(person[6])
        if person[4] == 'yes':
            ne_smoker_count += 1
        
sw_bmi_avg = round(sw_bmi_total/sw_count,2); sw_smoker_perc = round(sw_smoker_count/sw_count * 100,2); sw_ins_avg = round(sw_ins_cost/sw_count,2)
se_bmi_avg = round(se_bmi_total/se_count,2); se_smoker_perc = round(se_smoker_count/se_count * 100,2); se_ins_avg = round(se_ins_cost/se_count,2)
ne_bmi_avg = round(ne_bmi_total/ne_count,2); ne_smoker_perc = round(ne_smoker_count/ne_count * 100,2); ne_ins_avg = round(ne_ins_cost/ne_count,2)
nw_bmi_avg = round(nw_bmi_total/nw_count,2); nw_smoker_perc = round(nw_smoker_count/nw_count * 100,2); nw_ins_avg = round(nw_ins_cost/nw_count,2)

print('Southwest BMI Average | Smoker Percentage | Avg Insurance Cost: ' + str(sw_bmi_avg) + ' | ' + str(sw_smoker_perc) + ' | ' + str(sw_ins_avg))
print('Southeast BMI Average | Smoker Percentage | Avg Insurance Cost: ' + str(se_bmi_avg) + ' | ' + str(se_smoker_perc) + ' | ' + str(se_ins_avg))
print('Northeast BMI Average | Smoker Percentage | Avg Insurance Cost: ' + str(ne_bmi_avg) + ' | ' + str(ne_smoker_perc) + ' | ' + str(ne_ins_avg))
print('Northwest BMI Average | Smoker Percentage | Avg Insurance Cost: ' + str(nw_bmi_avg) + ' | ' + str(nw_smoker_perc) + ' | ' + str(nw_ins_avg))
```

```python
age_list = []
sex_list = []
bmi_list = []
children_list = []
smoker_list = []
cost_list = []

for person in insurance_data:
    age_list.append(float(person[0]))
    sex_list.append(person[1])
    bmi_list.append(float(person[2]))
    children_list.append(float(person[3]))
    smoker_list.append(person[4])
    cost_list.append(float(person[6]))

plt.scatter(age_list,cost_list)
plt.figure()
plt.bar(sex_list,cost_list)
plt.figure()
plt.scatter(bmi_list,cost_list)
plt.figure()
plt.scatter(children_list,cost_list)
plt.figure()
plt.bar(smoker_list,cost_list)
```
